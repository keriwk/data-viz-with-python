```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
```

```python
min_year = df['year'].min()
print(min_year)
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.
df_tot = df['year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
df_tot.plot(kind='bar', legend=False, color='blue')

plt.title('Total Books Published by Year', color='black')
plt.xticks(color='black')
plt.yticks(color='black')
plt.xlabel('Year', color='black')
plt.ylabel('Number of Books', color='black')
plt.savefig('bar_vertical.png')
plt.show()
```

```python
# Use your Python skills to create a visualization here.
plt.figure(figsize=(14, 6))
sb.barplot(x=df_tot.index, y=df_tot.values, color='blue')
plt.title('Total Books Published by Year', color='black')
plt.xticks(color='black')
plt.yticks(color='black')
plt.xlabel('Year', color='black')
plt.ylabel('Number of Books', color='black')
plt.show()
```
